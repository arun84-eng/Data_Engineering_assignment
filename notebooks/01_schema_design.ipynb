```python
con.execute("""
SELECT event_type, COUNT(*) as count
FROM read_csv_auto('C:/Users/arunb/Downloads/2019-Oct.csv')
GROUP BY event_type
""").fetchdf()
```

```python
con.execute("""
SELECT brand, COUNT(*) as count
FROM read_csv_auto('C:/Users/arunb/Downloads/2019-Oct.csv')
WHERE brand IS NOT NULL
GROUP BY brand
ORDER BY count DESC
LIMIT 10
""").fetchdf()
```
